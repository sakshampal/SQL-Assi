Q1. What is a database? Differentiate between SQL and NoSQL databases.

```python
Ans: A database functions, as a virtual storage unit like a filing cabinet,
    where one can effectively store and manage amounts of organized information. Think of it as an assortment
     of tables each dedicated to storing kinds of data such, as customer details, product information or financial records. 

        
         Differentiate between SQL and NoSQL databases.
        SQL
      .  SQL databases are primarily called Relational Databases (RDBMS)
    
      . SQL databases use a language called SQL to manage and manipulate data.
        SQL is a powerful tool for querying, inserting, updating, 
        and deleting data in a structured way.
        
      . In almost all situations SQL databases are vertically scalable. 
        This means that you can increase the load on a single server by
        increasing things like RAM, CPU, or SSD.
        
      .  SQL databases are usually relational, meaning they use tables to store data.
         Each table has rows (records) and columns (fields),
         and they are related to each other in a structured manner.
        
      . SQL databases are known for their strong consistency and reliability. 
        They follow the ACID (Atomicity, Consistency, Isolation, Durability)
        properties to ensure data integrity, even in high-stress situations.
        
        
        
        NOSQL (Not Only SQL)
        
      . NoSQL databases encompass various database technologies that don't
        strictly adhere to the traditional SQL approach. They provide
        more flexibility in data handling.
        
      . NoSQL databases can be document-based (e.g., MongoDB), key-value stores (e.g., Redis), 
        column-family stores (e.g., Cassandra), or graph databases (e.g., Neo4j). 
        This diversity allows them to handle different data types and structures.
        
      .  NoSQL databases are horizontally scalable. This means that you handle more
         traffic by sharding, or adding more servers in your NoSQL database. 

      . NoSQL databases are schema-less, meaning you can insert data without a predefined schema. 
         This flexibility is useful when dealing
        with rapidly changing or unstructured data. 
        
      . NoSQL databases are suitable for scenarios where data structures are dynamic,
        data volumes are high, and rapid scalability is needed, 
        such as social media platforms, real-time analytics, and IoT applications.
        
             
```

Q2.What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

```python
Ans: DDL(Data Definition Language): commands in databases are like everyday actions
    in the real world, such as building something new, throwing something away, changing
     something you already have, or clearing a table, 
     but applied to managing your database.
    
    CREATE: It's like building something new. When you use CREATE in databases, you're
    essentially telling it, "I want to make a new thing."
    For example, if you want to create a new list to keep track of your favorite movies,
    you're using the CREATE command in your database.
    
    CREATE TABLE Customers (
    ID INT PRIMARY KEY,
    Name VARCHAR(50),
    Email VARCHAR(100)
        
);
    
    DROP: Now, let's say you decide that you don't need that new room you just built in your house.
    Using DROP in SQL is like demolishing that
    room. It completely removes something from your database. For example:
        
     DROP TABLE Customers;
    
    ALTER: Think of ALTER as making changes to your house. If you want to add a window or a door to an
    existing room, you're altering it.
    Similarly, in SQL, you use ALTER to modify the structure of an existing database object. For example:
        
        ALTER TABLE Customers
        ADD Phone VARCHAR(15);
        
    TRUNCATE: Picture your bookshelf full of books. If you want to remove all the books without getting rid 
    of the shelf itself, you would "truncate" the shelf. In SQL, TRUNCATE is used to remove all
     the rows from a table while keeping the table's structure intact. 
        
    For example:
        TRUNCATE TABLE Customers;



    
```

```python
Q3.What is DML? Explain INSERT, UPDATE, and DELETE with an example.

Ans: DML (Data Manipulation Language): DML stands for Data Manipulation Language, and it's a
     set of SQL commands used for interacting with and  changing data in a database.

    
    INSERT: Imagine you have a shopping list, and you want to add a new item to it. Using
    INSERT in SQL is like adding that new item to your list.
    It's a way to put new data into your database.
    
    Example: You use INSERT to add "Milk" to your shopping list:

    INSERT INTO ShoppingList (Item) VALUES ('Milk');
    
    UPDATE: Now, suppose you've already written "Milk" on your shopping list, but you want to 
    change it to "Almond Milk" because that's what you actually need. Using UPDATE
    in SQL is like editing an item on your list.

    Example: You use UPDATE to change "Milk" to "Almond Milk":
    
    UPDATE ShoppingList SET Item = 'Almond Milk' WHERE Item = 'Milk';
    
    
    DELETE: Finally, let's say you've bought everything on your shopping list,
    and you want to remove all the items that are no longer needed. 
    Using DELETE in SQL is like crossing out or removing items from your list.

    Example: You use DELETE to remove all items from your shopping list that you've already bought:
    
    DELETE FROM ShoppingList WHERE Bought = 'Yes';




```

```python
Q4.What is DQL? Explain SELECT with an example.

Ans: DQL (Data Query Language): DQL stands for Data Query Language, and it's
     a subset of SQL (Structured Query Language) used for retrieving  information from a database. 

SELECT: When you use SELECT in SQL, it's like going to a library and asking the librarian to find
a particular book for you. You specify what you're looking for, and the database returns the relevant information.

Example: Imagine you have a library database, and you want to find all the books written by your favorite
author, Jane Doe. You would use SELECT in SQL like this:

SELECT Title, Author, PublicationYear
FROM Library
WHERE Author = 'Jane Doe';
```
